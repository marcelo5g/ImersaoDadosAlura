# Aula 02 - Análise Exploratória

```
# Selecionar todos os alunos com Idade igual a 13 anos (Não há nenhum do CEARÁ)
dados.query("NU_IDADE == 13")
```

```
# Selecionar pessoas com menos de 14 anos
dados.query("NU_IDADE <= 14")["SG_UF_RESIDENCIA"].value_counts()
```

```
# Calcular a proporção de participação de cada um dos Estados
# normalize = Colocar em valores entre 0 e 1
# A soma de todos os valores resulta em 1 (Proporção relativa)
dados.query("NU_IDADE <= 14")["SG_UF_RESIDENCIA"].value_counts(normalize=True)
```

```
alunos_menores_quatorze = dados.query("NU_IDADE <= 14")
```

```
#Gerar um gráfico para visualizar proporções.
#Gráfico de Pizza = Porcaria de acordo com Guilherme Silveira
alunos_menores_quatorze["SG_UF_RESIDENCIA"].value_counts().plot.pie(figsize=(10, 8))
```

```
#Gráfico de barras é melhor para visualizar
alunos_menores_quatorze["SG_UF_RESIDENCIA"].value_counts().plot.bar(figsize=(10, 8))
```

```
# Q006 = Questionário socioeconomico e Nota de Matemática
# seaborn = Biblioteca com gráficos com mais funcionalidades
import seaborn as sns

sns.boxplot(x="Q006", y="NU_NOTA_MT", data = dados)
```

```
# matplotlib = Permite acessar configurações especificas do seaborn como Título, Label Vertical e Horizontal
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.title("Boxplot das notas de matemática pela renda")
sns.boxplot(x="Q006", y="NU_NOTA_MT", data = dados)
```

```
# O gráfico acima não está ordenado no Eixo X. Preciso ordenar o Q006
# Sendo A = Renda Baixa e Q = Renda Alta
renda_ordenada = dados["Q006"].unique()
renda_ordenada.sort()
renda_ordenada
```

```
# Plotar novamente o gráfico, inserindo como argumento para order a renda_ordenada
plt.figure(figsize=(10, 8))
plt.title("Boxplot das notas de matemática pela renda")
sns.boxplot(x="Q006", y="NU_NOTA_MT", data = dados, order = renda_ordenada)
```

```
# O gráfico acima refere-se à nota de matemática. Como verificar uma nota total? Geral?
# Resposta = Somar as notas de todas as provas
```

```
# Esse comando não é o que eu quero, pois ele soma as colunas e eu quero as linhas
dados[provas].sum()
```

```
# Passo como argumento axis=1 (0 é a coluna e 1 é a linha)
dados[provas].sum(axis=1)
```

```
#Adicionar essa nova coluna com os totais das notas ao Dataframe
dados["NU_NOTA_TOTAL"] = dados[provas].sum(axis=1)
dados.head()
```

```
#Inserir a coluna criada como parametro para o eixo y do gráfic
plt.figure(figsize=(10, 8))
plt.title("Boxplot da Nota Total pela renda")
sns.boxplot(x="Q006", y="NU_NOTA_TOTAL", data = dados, order = renda_ordenada)
```

```
#Plotar um histograma com as notas totais a partir do Seaborn
sns.displot(dados, x="NU_NOTA_TOTAL")
```

```
# Olhar somente as pessoas que tiraram Zero
provas.append("NU_NOTA_TOTAL")
dados[provas].query("NU_NOTA_TOTAL == 0")
```

```
# Separar os dados somente das pessoas que fizeram pelo menos 1 dia de prova
# Ou seja, retirar do Dataset quem não compareceu em nenhum dos dias
dados_sem_notas_zero = dados.query("NU_NOTA_TOTAL != 0")
dados_sem_notas_zero.head()
```

```
plt.figure(figsize=(10, 8))
plt.title("Boxplot da Nota Total pela renda")
sns.boxplot(x="Q006", y="NU_NOTA_TOTAL", data = dados_sem_notas_zero, order = renda_ordenada)
```

```
# Adiciono um parametro - "hue" (Matiz, Tonalidade) para destacar essa caracteristica tambem
plt.figure(figsize=(14, 8))
plt.title("Boxplot da Nota Total pela renda")
sns.boxplot(x="Q006", y="NU_NOTA_TOTAL", data = dados_sem_notas_zero, 
            hue = "IN_TREINEIRO", order = renda_ordenada)
```

# Aula 03 - Distribuições e Correlações entre Notas

```
#Plotar um histograma somente com as notas totais (sem as notas Zero)
plt.figure(figsize=(12, 8))
sns.histplot(dados_sem_notas_zero, x="NU_NOTA_TOTAL")
```

```
#Plotar um histograma somente com as notas de matemática (sem as notas Zero)
# Percebe-se que a quantidade de pessoas que tiraram acima de 800 é bem baixa
plt.figure(figsize=(12, 8))
sns.histplot(dados_sem_notas_zero, x="NU_NOTA_MT")
```

```
#Plotar um histograma somente com as notas de Linguagens e Códigos (sem as notas Zero)
plt.figure(figsize=(12, 8))
sns.histplot(dados_sem_notas_zero, x="NU_NOTA_LC")
```

```
#Plotar um histograma somente com as notas totais (sem as notas Zero) e o questionario Q025 (Tem ou nao internet)
# B=Tem Internet / A = NÃO tem internet
# kde=True - Aproxima a distribuição 
plt.figure(figsize=(12, 8))
sns.histplot(dados_sem_notas_zero, x="NU_NOTA_TOTAL", hue="Q025", kde=True)
```

```
#Plotar um histograma somente com as notas totais (sem as notas Zero) e o questionario Q025 (Tem ou nao internet)
# B=Tem Internet / A = NÃO tem internet
# kde=True - Aproxima a distribuição 
# stat="density" - Normalizar pela área do histograma para que seja = a 1 (substitui o Count)
plt.figure(figsize=(12, 8))
sns.histplot(dados_sem_notas_zero, x="NU_NOTA_TOTAL", hue="Q025", kde=True, stat="density")
```

```
#Plotar um histograma somente com as notas totais (sem as notas Zero) e o questionario Q025 (Tem ou nao internet)
# B=Tem Internet / A = NÃO tem internet
# kde=True - Aproxima a distribuição 
# stat="density" - Normalizar pela área do histograma (substitui o Count)
# cumulative=True - 
plt.figure(figsize=(12, 8))
sns.histplot(dados_sem_notas_zero, x="NU_NOTA_TOTAL", hue="Q025", kde=True, stat="density", cumulative=True)
```

```
# Scatterplot
# Comparar notas de MT e LC (uma em função da outra)
# Parece existir uma relação (É esperado q quem estudou, estudou todas as matérias)
# Better Visualization = Quadrado(10, 10) e xlim e ylim com mesmos limites (do -50 até o 1050)
plt.figure(figsize=(10, 10))
sns.scatterplot(data= dados_sem_notas_zero, x="NU_NOTA_MT", y="NU_NOTA_LC")
plt.xlim((-50, 1050))
plt.ylim((-50, 1050))
```

```
# Pairplot
sns.pairplot(dados_sem_notas_zero[provas])
```

```
# Calcular a correlação em números (not only no gráfico)
# Correlação = Uma medida que vai de -1 até +1
# 0 = Nenhuma correlação / -1 = correlação inversamente proporcional forte (qdo uma sobe, a outra desce) 
# 1 = correlação diretamente proporcional forte (quando uma sobe, a outra sobe na mesma proporção)
correlacao = dados_sem_notas_zero[provas].corr()
correlacao
```

```
# Mapa de calor para analisar as correlações mais ou menos fortes
sns.heatmap(correlacao, cmap="Blues", center=0, annot=True)
```
