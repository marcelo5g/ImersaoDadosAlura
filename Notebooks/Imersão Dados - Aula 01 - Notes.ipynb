Como retornar apenas os alunos que são treineiros? (Na coluna, treineiro = 1, Ñ-Treineiro = 0)

```
dados.query("IN_TREINEIRO == 1")
```

Nessa query (filtro) só de treineiros, Vou selecionar A coluna Idade e contar quantas pessoas há em cada idade

```
dados.query("IN_TREINEIRO == 1")["NU_IDADE"].value_counts()
```

```
# lista com os nomes das colunas referentes as notas das provas
provas = ["NU_NOTA_CH", "NU_NOTA_MT", "NU_NOTA_LC", "NU_NOTA_CN","NU_NOTA_REDACAO"]
```

```
dados[provas].describe()
```

Quantil = Qual a nota mínima dos 10% com as maiores notas?

```
dados["NU_NOTA_LC"].quantile(0.9)
```

Como representar o .describe( ) na forma de um gráfico? 
O Retângulo representa a faixa onde se concentram a maioria das notas
A Linha (meio do retangulo) representa a mediana
Os Pontos são os outliers

```
dados["NU_NOTA_LC"].plot.box(grid = True, figsize=(8,6))
```

```
# gera um gráfico boxplot de todas notas
dados[provas].boxplot(grid=True, figsize=(8, 6))
```
