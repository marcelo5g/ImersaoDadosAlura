# Aula 04 - Machine Learning

Não estamos avaliando classes específicas de novas mas o Intervalo de valores de notas. >>> Problema de Regressão

Qual o problema que eu quero resolver? 
R: Será que eu consigo prever a nota de matemática baseado nas outras avaliações?


```
provas_entrada = ["NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_CN","NU_NOTA_REDACAO"]
provas_saida = "NU_NOTA_MT"

# Remover os NaN do datase dados_sem_notas_zero
dados_sem_notas_zero = dados_sem_notas_zero[provas].dropna()

notas_entrada = dados_sem_notas_zero[provas_entrada]
notas_saida = dados_sem_notas_zero[provas_saida]
```

```
notas_entrada
```

```
x = notas_entrada
y = notas_saida
```

```
# Preciso separar dados para TREINO e dados para TESTE
# test-size=0.25 - 25% dos dados separados para TESTE
# SEED = variavel com valor fixo(4321) para garantir que o random seja sempre igual?
from sklearn.model_selection import train_test_split

SEED = 4321

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.25,
                                                        random_state=SEED)
```

```
x_treino.head()
```

```
x_treino.shape 
```

```
# Valores devem ser iguais para X e para Y
len(y_treino)
```

```
# Criar um modelo de regressão linear
# Solicitar ao modelo de ML que se ajuste (FIT) aos dados de treino
# LinearSVR é uma variação do SVM(Suport Virtual Machine) que é Relacionamentos Lineares
from sklearn.svm import LinearSVR

modelo = LinearSVR(random_state = SEED)
modelo.fit(x_treino, y_treino)
```

```
# Instrução ao modelo criado acima para, baseado em x_test, predizer as NOTAS_MT
modelo.predict(x_teste)
```

```
# Comparar o array com os valores reais com os resultados da previsão
y_teste[:5]
```

```
# Como avaliar se o modelo está bom ou se está ruim
# Guardo a predição numa variavel e ploto o scatterplot dela comparando com o scatterplot dos valores reais
predicoes_matematica = modelo.predict(x_teste)
```

```
# Se a prediçao acertasse tudo, todos os pontos iriam coincidir, formando uma única Linha reta (linear)
plt.figure(figsize=(10, 10))
sns.scatterplot(x=predicoes_matematica, y=y_teste)
plt.xlim((-50, 1050))
plt.ylim((-50, 1050))
```

```
# Eixo X = nota original da Pessoa, Eixo y = o quanto que eu errei (diferenca entre o real e a predicao)
# Exemplo: se o modelo previu 300 e a nota foi 275, o erro foi de 25 
plt.figure(figsize=(10, 10))
sns.scatterplot(x=y_teste, y=y_teste - predicoes_matematica)
```

```
# Aqui vamos correlacionar o valor real da prova de matemática com a média de todas as outras provas
plt.figure(figsize=(10, 10))
sns.scatterplot(x=y_teste, y=x_teste.mean(axis=1))
plt.xlim((-50, 1050))
plt.ylim((-50, 1050))
```

```
# Como visualizar a diferenca entre o real e a predicao por meio de numeros
# Criar um novo DataFrame chamado Resultados com os valores que preciso para comparar
# Quadrado da diferença = Elevo ao quadrado para eliminar os números negativos
resultados = pd.DataFrame()
resultados["Real"] = y_teste
resultados["Previsao"] = predicoes_matematica
resultados["diferenca"] = resultados["Real"] - resultados["Previsao"]
resultados["quadrado_diferenca"] = (resultados["Real"] - resultados["Previsao"])**2
```

```
# A média de todos os "quadrado_diferenca" resume como o Modelo performa no geral
resultados["quadrado_diferenca"].mean()
```

```
# Ao calcular a raiz quadrado, percebe-se que o modelo está errando em média 80 pontos para cima ou para baixo
resultados["quadrado_diferenca"].mean()**(1/2)
```

# Aula 05 - Validação de modelo e Overfit

```
# mean_squared_error = Avalia a modelo, calcula a taxa de erro (Varia muiro por conta o random) 
# Como diminui o fator de aleatoriedade do modelo?
from sklearn.svm import LinearSVR

modelo = LinearSVR(random_state = SEED)
modelo.fit(x_treino, y_treino)
predicoes_matematica = modelo.predict(x_teste)
mean_squared_error(y_teste, predicoes_matematica)
```

```
# Selecionar um modelo que dependa um pouco menos do 'random_state' 
# Modelo Decision Tree (Árvore de Decisão) que só usa o random em casos bem específicos (não é o nosso caso)
# Aqui, usando o modelo de árvore, consigo obter sempre o mesmo valor
from sklearn.tree import DecisionTreeRegressor

modelo_arvore = DecisionTreeRegressor(max_depth = 3)
modelo_arvore.fit(x_treino, y_treino)

predicoes_matematica_arvore = modelo_arvore.predict(x_teste)
mean_squared_error(y_teste, predicoes_matematica_arvore)
```

```
# Porém, se eu fizer uma nova atribuição para as variavels de X/Y treino/test
# Utilizando o train_test_split(isso tb gera aleatoriedade, pois ele seleciona 25% de forma aleatoria)
# 75% pra treino e 25% para teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.25)
modelo_arvore = DecisionTreeRegressor(max_depth = 3)
modelo_arvore.fit(x_treino, y_treino)

predicoes_matematica_arvore = modelo_arvore.predict(x_teste)
mean_squared_error(y_teste, predicoes_matematica_arvore)
```

```
# Para obter uma melhor performance, eu realizo esse processo várias vezes (repetição), escolhendo diferentes "25%s"
# Escolho inumeras possibilidades (Folds) de "25%s" e "tiro a média"
# cv = Quantas vezes irei realizar a repetição do processo (cross validate)
# Para fazer isso, utilizo o Cross_validate
# No scoring, configuro a mesma "unidade de medida de erro" ("Quanto maior, melhor" - se está negativo, está ruim)
# neg = O scoring(pontuação) é o contrario do erro

from sklearn.model_selection import cross_validate

modelo_arvore = DecisionTreeRegressor(max_depth=2)
resultados = cross_validate(modelo_arvore, x, y, cv = 5, scoring="neg_mean_squared_error")
resultados
```

```
resultados["test_score"]*(-1)
```

```
media = (resultados["test_score"]*(-1)).mean()
media
```

```
# Calcular o desvio padrão e o intervalo de confiança (Limites superiores e inferiores)
desvio_padrao = (resultados["test_score"]*(-1)).std()
desvio_padrao
```

```
lim_inferior = media - (2*desvio_padrao)
lim_superior = media + (2*desvio_padrao)

print(f'Intervalo de confiança {lim_inferior} até {lim_superior}')
```

```
def calcula_mse(resultados):
    media = (resultados["test_score"]*(-1)).mean()
    desvio_padrao = (resultados["test_score"]*(-1)).std()
    lim_inferior = media - (2*desvio_padrao)
    lim_superior = media + (2*desvio_padrao)

    print(f'Intervalo de confiança {lim_inferior} até {lim_superior}')
```

```
# KFold = Embaralha os dados (garantir que não irá utilizar uma lista ordenada e pegar sempre os mesmos dados)
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold

partes = KFold(n_splits = 10, shuffle=True)
modelo_arvore = DecisionTreeRegressor(max_depth=2)
resultados = cross_validate(modelo_arvore, x, y, cv = partes, scoring="neg_mean_squared_error")
calcula_mse(resultados)
```

```
# KFold = Embaralha os dados (garantir que não irá utilizar uma lista ordenada e pegar sempre os mesmos dados)
# uso o numpy e o SEED para obter sempre os mesmos dados (já embaralhados) para poder comparar resultados 
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
import numpy as np

SEED = 1232
np.random.seed(SEED)

partes = KFold(n_splits = 10, shuffle=True)
modelo_arvore = DecisionTreeRegressor(max_depth=2)
resultados = cross_validate(modelo_arvore, x, y, cv = partes, scoring="neg_mean_squared_error")
calcula_mse(resultados)
```

```
# Quanto maior o max_depth, maior a profundidade da minha árvore de decisões e melhor o meu resultado

def regressor_arvore(nivel):
    SEED = 1232
    np.random.seed(SEED)
    partes = KFold(n_splits = 10, shuffle=True)
    modelo_arvore = DecisionTreeRegressor(max_depth=nivel)
    resultados = cross_validate(modelo_arvore, x, y, cv = partes, scoring="neg_mean_squared_error", return_train_score=True)
    print(f"Treino = {(resultados['train_score']*-1).mean()} | Teste = {(resultados['test_score']*-1).mean()}")

regressor_arvore(4)
```

```
# Lá pelo i=8 a tendencia se inverte e os numeros do Teste começçam a aumentar, gerando resultados piores
# Porém os números do Treino diminuem 
# Overfit = Quando os dados de Treino acabam quase decorando os dados mas não conseguindo prever resultados no Teste

for i in range(1, 21):
    regressor_arvore(i)
```
